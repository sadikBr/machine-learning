```python
#creating some utility functions
def sigmoid(z):
  return 1 / (1 + np.exp(-z))

def h(theta, x_data):
  return sigmoid(np.dot(x_data, theta.T))
```

```python
#creating the feed forward algorithm
def feed_forward(x_data, theta1, theta2):
  ones = np.ones((len(x_data), 1))
  x_data = np.hstack((ones, x_data))

  out = h(x_data, theta1)
  output = np.hstack((ones, out))

  predictions = h(output, theta2)

  return np.argmax(predictions, axis=1) + 1
```
