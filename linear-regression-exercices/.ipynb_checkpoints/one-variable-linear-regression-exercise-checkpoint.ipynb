```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read the csv file
data = pd.read_csv('ex1data1.txt', names=['Population', 'Profit'])

#spliting the features from the target
x_data = data["Population"].values
y_data = data["Profit"].values.reshape(x_data.size, 1)
```

```python
#normalizing the data
x_data = (x_data - x_data.mean()) / x_data.std()

#ploting the training data
plt.title('Scatte plot of training data.')
plt.ylabel('Profit in 10k')
plt.xlabel('Population of city')
plt.scatter(x_data, y_data, color='r', marker='x')
```

```python
#formating the x_data to be a matrix of shape (m, 2)
ones = np.ones((x_data.size, 2))
ones[:,1] = x_data

x_data = ones
```

```python
#creating utility functions

#the cost function J(theta) = (1 / 2*m) * sum(h_theta(x) - y)^2
def cost_function(x_data, y_data, theta):
    h = np.dot(x_data, theta)
    error_sq = (h - y_data) ** 2
    error_sum = np.sum(error_sq)
    
    return (1 / (2 * x_data.size)) * error_sum

#gradient descent function
def gradient_descent(x_data, y_data, theta, learning_rate):
    h = np.dot(x_data, theta)
    error = h - y_data
    dtheta = (1 / x_data.size) * np.dot(x_data.T, error)
    theta = theta - learning_rate * dtheta
    
    return theta
```

```python
#training function
def train_model(x_data, y_data, learning_rate, nbr_iter):
    theta = np.random.randn(2, 1)
    cost = []
    for i in range(nbr_iter):
        cost.append(cost_function(x_data, y_data, theta))
        theta = gradient_descent(x_data, y_data, theta, learning_rate)
        
    return theta, cost
```

```python
#training the model
learning_rate = 0.4
nbr_iter = 50

theta, cost = train_model(x_data, y_data, learning_rate, nbr_iter)
# cost_function(x_data, y_data, theta)
# gradient_descent(x_data, y_data, theta, learning_rate)
```

```python
#ploting the cost function over time
x = np.arange(nbr_iter)
plt.plot(x, cost)
```

```python
#ploting the linear regression model
plt.scatter(x_data[:, 1], y_data, color='blue', marker='+')
plt.plot(x_data[:, 1], np.dot(x_data, theta), color='red')
plt.title('Linear Regression trained model')
plt.xlabel('Population of city')
plt.ylabel('Profit')
```
