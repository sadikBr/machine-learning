```python
#importing the needed libraries
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
#importing the data and the weights
data = loadmat('ex3data1.mat')
weights = loadmat('ex3weights.mat')

x_data = data['X']
y_data = data['y']

theta1 = weights['Theta1']
theta2 = weights['Theta2']
```

```python
# plotting some of the input data (x_data)
fig, axis = plt.subplots(10,10,figsize=(5,5))
for i in range(10):
    for j in range(10):
        axis[i,j].imshow(x_data[np.random.randint(0,5001)].reshape(20,20,order="F"), cmap='Blues') #reshape back to 20 pixel by 20 pixel
        axis[i,j].axis("off")
```

```python
#creating some utility functions
def sigmoid(z):
  return 1 / (1 + np.exp(-z))

def h(x_data, theta):
  return sigmoid(np.dot(x_data, theta.T))
```

```python
#creating the feed forward algorithm
def feed_forward(x_data, theta1, theta2):
  ones = np.ones((len(x_data), 1))
  x_data = np.hstack((ones, x_data))

  out = h(x_data, theta1)
  output = np.hstack((ones, out))

  predictions = h(output, theta2)

  return np.argmax(predictions, axis=1) + 1
```

```python
#calculating the probabilities
predictions = feed_forward(x_data, theta1, theta2).reshape(len(y_data), 1)
```

```python
#calculate the precision of the model
def precision(y_data, predictions):
  return str((np.sum(y_data == predictions) / len(y_data)) * 100)+'%'
```

```python
precision(y_data, predictions)
```
