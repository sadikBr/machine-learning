```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating the data

x_data = np.arange(250, 700)

def fn(x):
  return 1.7 * x + 2.5

y_data = fn(x_data) + np.random.randn(x_data.size) * np.random.randint(50, 500)
```

```python
np.random.randint(50, 500,(1,2))
```

```python
#visualisation of the data

plt.scatter(x_data, y_data)
plt.show()
```

```python
#data normalization

x_data = (x_data - x_data.mean()) / (x_data.std())
```

```python
#data shuffling

np.random.shuffle(x_data)
```

```python
#spliting the data

training_per = int(x_data.size * 0.8)

x_train = x_data[:training_per]
y_train = y_data[:training_per].reshape(training_per, 1)

x_test = x_data[training_per:]
y_test = y_data[training_per:].reshape(x_data.size - training_per, 1)

x_train.size, y_train.size, x_test.size, y_test.size
```

```python
ones_train = np.ones((x_train.size, 2))
ones_test = np.ones((x_test.size, 2))

ones_train[:, 1] = x_train
ones_test[:, 1] = x_test
```
